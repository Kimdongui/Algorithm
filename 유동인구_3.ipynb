```python
def make_plot_age(HD):
    fig, axes = plt.subplots(nrows=5,ncols=4)
    fig.set_size_inches(12, 18)
    age = 0
    x = 0
    y = 0
    
    while age <= 14:
        for x in range(4):
            for y in range(4):
                if x == 3 & y == 3:
                    continue
                axes[x][y].plot(HD[HD.columns[1]], HD[HD.columns[age+4]])
                axes[x][y].plot(HD[HD.columns[1]], HD[HD.columns[age+19]])
                age += 1
    axes[4][0].plot(HD[HD.columns[1]], HD[HD.columns[34]])
    axes[4][0].plot(HD[HD.columns[1]], HD[HD.columns[35]])
    axes[4][1].plot(HD[HD.columns[1]], HD[HD.columns[36]], c='black')
    
    plt.xlim(0, 30)
    plt.ylim(0, 1000000)
    mpl.rc('font', family='NanumGothic')
    plt.suptitle(HD.HDONG_NM.values[0])
    plt.show()

make_plot_age(HDONG_0)
```

```python
def make_plot_HDONG():
    fig, axes = plt.subplots(nrows=10,ncols=7)
    fig.set_size_inches(12, 30)
    idx_HD = 0
    x = 0
    y = 0
#     HD = 'HDONG_{}'.format(i)
    HD = []
    
    for index in range(0,1932,day_of_month):
        globals()['HDONG_{}'.format(code)] = train_df[index:index+day_of_month]
        code += 1
    
    for i in range(69):
        HD.append('HDONG_{}'.format(i))
        
    while idx_HD < 69:
        for x in range(10):
            for y in range(7):
                axes[x][y].plot(HD[idx_HD][HD.columns[1]], HD[HD.columns[34]]) # 남 나이대 합
                axes[x][y].plot(HD[HD.columns[1]], HD[HD.columns[35]]) # 여 나이대 합
                axes[x][y].plot(HD[HD.columns[1]], HD[HD.columns[36]]) # 총 합

                age += 1
    
    plt.xlim(0, 30)
    plt.ylim(0, 1000000)
    mpl.rc('font', family='NanumGothic')
    plt.suptitle(HD.HDONG_NM.values[0])
    plt.show()

make_plot_HDONG()
```
