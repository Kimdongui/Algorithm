```python
import datetime

def make_weekday(x):
    
    week = ['월','화','수','목','금','토','일']

    day = datetime.datetime.strptime(str(x),'%Y%m%d').weekday()
    
    return week[day]

make_weekday(20190201)

```

```python
Flow_Time_1902['STD_DAY'] = Flow_Time_1902["STD_YMD"].map(make_weekday)
Flow_Time_1903['STD_DAY'] = Flow_Time_1903["STD_YMD"].map(make_weekday)
Flow_Time_1904['STD_DAY'] = Flow_Time_1904["STD_YMD"].map(make_weekday)
Flow_Time_1905['STD_DAY'] = Flow_Time_1905["STD_YMD"].map(make_weekday)

Flow_Time_2002['STD_DAY'] = Flow_Time_2002["STD_YMD"].map(make_weekday)
Flow_Time_2003['STD_DAY'] = Flow_Time_2003["STD_YMD"].map(make_weekday)
Flow_Time_2004['STD_DAY'] = Flow_Time_2004["STD_YMD"].map(make_weekday)
Flow_Time_2005['STD_DAY'] = Flow_Time_2005["STD_YMD"].map(make_weekday)

Flow_Time_2002
```

```python
# 8x4 형식의 그래프 만들기.
# x 축 : 시간대
# y 축 : 유동인구 수
# 각 특정 동에 대해서 나타내는 전체 표
# 한 그래프 당 일자 별을 나타냄

def Make_Time_HDONG(month, num_HD):
    
    HDONG = []
    
    if month == 1902:
        day = 28
        base_df = Flow_Time_1902
    elif month == 1903:
        day = 31
        base_df = Flow_Time_1903
    elif month == 1904:
        day = 30
        base_df = Flow_Time_1904
    elif month == 1905:
        day = 31
        base_df = Flow_Time_1905
    elif month == 2002:
        day = 29
        base_df = Flow_Time_2002
    elif month == 2003:
        day = 31
        base_df = Flow_Time_2003
    elif month == 2004:
        day = 30
        base_df = Flow_Time_2004
    elif month == 2005:
        day = 31
        base_df = Flow_Time_2005
    else:
        print("1902~1905, 2002~2005")
    num_hdong = 69
    sum_index = day * num_hdong
    
    base_df = base_df.sort_values(by=['HDONG_CD','STD_YMD'])
    base_df = base_df.reset_index()
    
    for index in range(0, sum_index, day):
        HDONG.append(base_df[index:index+day])
        
    return HDONG[num_HD]
```

```python
# 8x4 형식의 그래프 만들기.
# x 축 : 시간대
# y 축 : 유동인구 수
# 각 특정 동에 대해서 나타내는 전체 표
# 한 그래프 당 일자 별을 나타냄

import copy

def Make_Time_Plot_HDONG(HD):
    fig, axes = plt.subplots(nrows=8,ncols=4)
    fig.set_size_inches(12, 20)
    day = 0
    x = 0
    y = 0
    mpl.rc('font', family='NanumGothic')
    
    base_df = copy.copy(HD)
    
    del base_df['index']
    base_df = base_df.reset_index(drop=True)
    del base_df['STD_YM']
    del base_df['STD_YMD']
    del base_df['HDONG_CD']
    del base_df['HDONG_NM']
    del base_df['STD_DAY']
    del base_df['TMST_SUM']
    
    base_df = base_df.T
    base_df = base_df.reset_index(drop=True)
    y_limit = [30000 for i in range(len(base_df.index))]
    
    while day <= 31:
        for x in range(8):
            for y in range(4):
                if day in base_df.columns:
                    if HD['STD_DAY'].values[day] == '토' or HD['STD_DAY'].values[day] == '일':
                        axes[x][y].plot(base_df.index, base_df[base_df.columns[day]], color='r')
                    elif HD['STD_DAY'].values[day] == '금':
                        axes[x][y].plot(base_df.index, base_df[base_df.columns[day]], color='orange')
                    else:
                        axes[x][y].plot(base_df.index, base_df[base_df.columns[day]], color='b')
                    axes[x][y].plot(base_df.index, y_limit)
                    axes[x][y].set_title(str(HD['STD_YMD'].values[day])+' '+str(HD['STD_DAY'].values[day]))
                    day += 1
                else:
                    day += 1
                    break
    
    plt.suptitle(str(HD.HDONG_NM.values[0]))
    plt.show()
```

```python
ex = Make_Time_HDONG(1902, 0)
Make_Time_Plot_HDONG(ex)
```

```python
# 8x4 형식의 그래프 만들기.
# x 축 : 시간대
# y 축 : 유동인구 수
# 각 일자에 대해서 나타내는 전체 표
# 한 그래프 당 특정 동을 나타냄

def Make_Time_Plot_Day(FullDay):
    fig, axes = plt.subplots(nrows=10,ncols=7)
    fig.set_size_inches(12, 30)
    x = 0
    y = 0
    mpl.rc('font', family='NanumGothic')
    
    if FullDay // 100 == 201902:
        base_df = copy.copy(Flow_Time_1902)
    elif FullDay // 100 == 201903:
        base_df = copy.copy(Flow_Time_1903)
    elif FullDay // 100 == 201904:
        base_df = copy.copy(Flow_Time_1904)
    elif FullDay // 100 == 201905:
        base_df = copy.copy(Flow_Time_1905)
    elif FullDay // 100 == 202002:
        base_df = copy.copy(Flow_Time_2002)
    elif FullDay // 100 == 202003:
        base_df = copy.copy(Flow_Time_2003)
    elif FullDay // 100 == 202004:
        base_df = copy.copy(Flow_Time_2004)
    elif FullDay // 100 == 202005:
        base_df = copy.copy(Flow_Time_2005)
    else:
        print('20190201 ~ 20200531')
        
    ori_df = copy.copy(base_df)
    del base_df['STD_YM']
    del base_df['STD_YMD']
    del base_df['HDONG_CD']
    del base_df['HDONG_NM']
    del base_df['STD_DAY']
    del base_df['TMST_SUM']
    
    time_list = [i for i in range(24)]
    y_limit = [10000 for i in range(len(time_list))]
    
    day = FullDay % 100
    # 1~31일 . 1일: 0~69. 2일: 69~138. ~~ ,, 행정동 갯수가 69개.
    day_HDONG = day*69 - 69
    end_day_HDONG = day*69
    
    while day_HDONG < end_day_HDONG:
        for x in range(10):
            for y in range(7):
                if day_HDONG >= end_day_HDONG:
                    break
                axes[x][y].plot(time_list, base_df.loc[day_HDONG])
                axes[x][y].plot(time_list, y_limit)
                axes[x][y].set_title(ori_df.HDONG_NM.values[day_HDONG])
                day_HDONG += 1
    
    plt.suptitle(str(FullDay)+' '+str(ori_df['STD_DAY'].values[day_HDONG-2]))
    plt.show()
    print(day_HDONG)
```
