```python
newdataT = newdata.T
newdataT.head()
```

```python
n_data = newdataT.loc[[4,20,40,48],list(newdataT.columns)]
n_data.head()
```

```python
del n_data["행정동코드"]
```

```python
n_data.T.head()
```

```python
data_2019_02 = n_data.T[3:31].mean()
data_2019_03 = n_data.T[31:62].mean()
data_2019_04 = n_data.T[62:92].mean()
data_2019_05 = n_data.T[92:123].mean()
data_2020_02 = n_data.T[123:152].mean()
data_2020_03 = n_data.T[152:183].mean()
data_2020_04 = n_data.T[183:213].mean()
data_2020_05 = n_data.T[213:244].mean()
```

```python
gu_data =pd.concat([data_2019_02, data_2019_03, data_2019_04, data_2019_05, data_2020_02, data_2020_03, data_2020_04, data_2020_05],axis=1)
```

```python
gu_data.columns = ["201902", "201903", "201904", "201905", "202002", "202003", "202004", "202005"]
```

```python
gu_data.index = ["서울 중구", "서울 노원구", "대구 중구", "대구 수성구"]
```

```python
gu_data.head()
```

```python
gu_dataT = gu_data.T
gu_dataT
```

```python
data_2019=gu_dataT[:4]
```

```python
data_2020=gu_dataT[4:]
```

```python
X = np.arange(4)
fig = plt.figure(figsize = (18,20))
fig.suptitle("매출지수",fontsize= 24)
for i in range(0,4):
    plt.subplot(4,4,i+1)
    plt.bar(X+2.00, list(data_2019[data_2019.columns[i]]), label='2019년',color = 'b', width = 0.25)
    plt.bar(X+2.25, list(data_2020[data_2020.columns[i]]), label='2020년',color = 'r', width = 0.25)
    plt.legend(loc="best",frameon=False)
    plt.title(gu_dataT.columns[i])
    plt.xlabel("월")
    plt.ylabel("%(퍼센트)")
plt.show()
```
